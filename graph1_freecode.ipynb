```python
# has path

from collections import deque
graph = {
    "f" : ["g","i"],
    "g" : ["h"],
    "h" : [],
    "i" : ["g", "k"],
    "j" : ["i"],
    "k" : []
}


def has_path(graph, src, des):

    if src == des:
        return True
    
    for neighbor in graph[src]:
        if has_path(graph, neighbor, des):
            return True

    return False

print(has_path(graph, 'f', 'j'))


def has_path_bfs(graph, src, des):

    queue = deque()

    queue.append(src)

    while queue:

        cur = queue.popleft()

        if cur == des:
            return True

        for neighbor in graph[cur]:
            queue.append(neighbor)
    
    return False

print(has_path_bfs(graph, 'f', 'j'))
```

```python
# undirected path
graph = {
    'i': ['j', 'k'],
    'j': ['i'],
    'k': ['i', 'm', 'l'],
    'm': ['k'],
    'l': ['k'],
    'o': ['n'],
    'n': ['o']
}


def has_path_cyclic(graph, src, des, visited = set()):

    if src ==  des:
        return True

    visited.add(src)

    for neighbor in graph[src]:
        if neighbor not in visited and has_path_cyclic(graph, neighbor, des, visited):
            return True
    
    return False
    

print(has_path_cyclic(graph, 'j', 'o'))
```

```python
# number of connected component in undirected graph

graph = {
    3: [],
    4: [6],
    6: [4, 5, 7, 8],
    8: [6],
    7: [6],
    5: [6],
    1: [2],
    2: [1]
}


def numberOfConnectedComp(graph):
    
    visited = set()

    def dfs(graph, src, visited):
        
        if src in visited:
            return False

        visited.add(src)

        for neighbor in graph[src]:
            if neighbor not in visited:
                dfs(graph, neighbor, visited)

        return True
        
    no_of_comp = 0

    for key in graph.keys():
        if dfs(graph, key, visited):
            no_of_comp += 1

    return no_of_comp


print(numberOfConnectedComp(graph))


    
```

```python
# shortest path
graph = {
    'w': ['x', 'v'],
    'x': ['w', 'y'],
    'y': ['x', 'z'],
    'z': ['y', 'v'],
    'v': ['z', 'w']
}

def find_shortest_path(graph, src, des):
    
    def bfs(graph, src, path=[], visited=set()):
        queue = deque()

        queue.append(src)

        while queue:
            cur = queue.popleft()
            path += [cur]
            if cur == des:
                return path
            if cur not in visited:
                visited.add(cur)       
                for neighbor in graph[cur]:
                    queue.append(neighbor)
            
        return []
    
    print(bfs(graph, 'w'))


find_shortest_path(graph, 'w', 'z')

```

```python
# binary search

# to find the index where we can insert the a number

arr = [3,5]

def bs(arr, num):
    l, r = 0, len(arr)-1

    while l < r:
        mid = (l+r)//2

        if arr[mid] < num:
            l = mid+1
        else:
            r = mid-1

    return l 

print(bs(arr, 5))
```

```python
from collections import deque

def cycleDetection(graph):
    visited = set()

    def bfs(start):
        queue = deque([(start, -1)])
        
        while queue:
            node, parent = queue.popleft()
            
            if node in visited:
                return True
            
            visited.add(node)
            
            for neighbor in graph[node]:
                if neighbor != parent:
                    queue.append((neighbor, node))
        
        return False

    for node in graph.keys():
        if node not in visited:
            if bfs(node):
                return True
    
    return False

# Example usage
graph = {
    0: [1, 2],
    1: [0, 2],
    2: [0, 1, 3],
    3: [2]
}

graph1 = {
    0: [1],
    1: [0],
    2: [3,4],
    3: [2,4],
    4: [2,3]
}

graph2 = {
    0: [1],
    1: [0,2],
    2: [1]
}

print(cycleDetection(graph))
print(cycleDetection(graph1))
print(cycleDetection(graph2))
```
